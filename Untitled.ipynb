```python
import cv2
import numpy as np
import base64
img_data = base64.b64decode(img)
nparr = np.frombuffer(img_data, np.uint8)
image = cv2.imdecode(nparr, cv2.IMREAD_COLOR)
```

```python
cv2.imshow("img",image)
```
